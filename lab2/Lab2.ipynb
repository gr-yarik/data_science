```python
import numpy as np
import pandas as pd
data = pd.read_excel("rollingsales_bronx.xls", sheet_name="Bronx")
data
```

```python
data = data[data["SALE&PRICE"] != 0]
data
```

```python
def is_year(year, mean):
    return mean if year == 0 else year

def is_int(number):
    return number if isinstance(number, np.integer) else 1
```

```python
mean_y = data["YEAR BUILT"].mean()

data["TAX CLASS AT TIME OF SALE"] = data["TAX CLASS AT TIME OF SALE"].apply(is_int)
data["YEAR BUILT"] = data["YEAR BUILT"].apply(lambda x: is_year(x, mean_y))
data
```

```python
data_for_plt = data.groupby("YEAR BUILT")[["SALE&PRICE"]].median()
data_for_plt.plot(xlabel="YEAR BUILT", ylabel="median of price", grid ="true")
```

```python
data_for_plt = data.groupby("YEAR BUILT")[["SALE&PRICE"]].mean()
data_for_plt.plot(xlabel="YEAR BUILT", ylabel="median of price", grid ="true")
```

```python
data_for_plt = data.groupby("BUILDING CLASS CATEGORY")[["SALE&PRICE"]].mean()
data_for_plt.plot(kind ="bar", xlabel="Building class category", ylabel="Mean of price")
```

```python
data_for_plt = data.groupby("NEIGHBORHOOD")[["SALE&PRICE"]].count()
data_for_plt.plot(kind ="bar", xlabel="NEIGHBORHOOD", ylabel="Number of sales")
```

```python
data_for_plt = data.groupby("BUILDING CLASS CATEGORY")[["SALE&PRICE"]].count()
data_for_plt.plot(kind ="bar", xlabel="Building class category", ylabel="Number of sales")
```

```python
data_for_plt = data.groupby([data["SALE DATE"].dt.to_period('M')])[["SALE&PRICE"]].count()
data_for_plt.plot(kind ="bar", xlabel="date", ylabel="number of sales")
```

```python
data_for_plt = data.groupby([data["SALE DATE"].dt.to_period('M')])[["SALE&PRICE"]].sum()
data_for_plt.plot(kind ="bar", xlabel="date", ylabel="sum sale")
```
